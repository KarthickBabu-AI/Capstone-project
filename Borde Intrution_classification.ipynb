```python
#Importing pandas library for the dataset
import pandas as pd 
```

```python
#Reading the file
dataset=pd.read_csv("border_intrusion_dataset.csv")
```

```python
#To view the dataset
dataset
```

```python
# To view Null values
dataset.isnull().sum()
```

```python
import numpy as np  # Imports NumPy to handle numerical data and represent missing values as np.nan
from sklearn.impute import SimpleImputer  # Loads SimpleImputer for filling missing values
imp_mean = SimpleImputer(missing_values=np.nan, strategy='most_frequent')  # Creates an imputer that replaces NaN with column median
imp_mean.fit(dataset[['weather_condition']])  # Learns mean values from all quantitative columns in the dataset
df_qual = imp_mean.transform(dataset[['weather_condition']])  # Applies mean replacement to missing values in those columns
df_qual=pd.DataFrame(df_qual,columns=['weather_condition'])
df_qual
```

```python
dataset.isnull().sum()
```

```python
#Using for loop to fill the null value
for col in ['weather_condition']:
    dataset.loc[:, col] = dataset[col].fillna(dataset[col].mode()[0])
```

```python
# To view Null values
dataset.isnull().sum()
```

```python
#Using get_dummies to convert text into number
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```
